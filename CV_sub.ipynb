```python
from PIL import Image
import os
from os import listdir
  
def convertThreatImage():
  folder = "./content/sample_data/threat_images"
  for x in os.listdir(folder):
    print(x)
    img = Image.open(folder+"/"+x)
    print(img)
    img = img.convert("RGBA")
  
    datas = img.getdata()
  
    newData = []
  
    for item in datas:
      if item[0] == 255 and item[1] == 255 and item[2] == 255:
          newData.append((255, 255, 255, 0))
      else:
          newData.append(item)
  
    img.putdata(newData)

    img.save("./content/sample_data/cropped_images/"+x.split(".")[0]+".png", "PNG")
    print("Successful")

def convertBackImages():
  folder = "./content/sample_data/background_images"
  for x in os.listdir(folder):
    print(x)
    img = Image.open(folder + "/"+x)
    print(img)
    img = img.convert("RGBA")
  
    datas = img.getdata()
  
    newData = []
  
    for item in datas:
      if item[0] == 255 and item[1] == 255 and item[2] == 255:
          newData.append((255, 255, 255, 0))
      else:
          newData.append(item)
  
    img.putdata(newData)
    img.save("./content/sample_data/croppedbg_images/"+x.split(".")[0]+".png", "PNG")
    print("Successful")

convertThreatImage()
convertBackImages()

folder1 = "./content/sample_data/croppedbg_images"
folder2 = "./content/sample_data/cropped_images"

for x in os.listdir(folder1):
  if x == '.ipynb_checkpoints':
    continue
  for y in os.listdir(folder2):
    img1 = Image.open(folder1 +"/" +x)
    img1copy = img1.copy().resize((500,500))
    bg_w, bg_h = img1copy.size
    if y == '.ipynb_checkpoints':
      continue
    img2 = Image.open(folder2 + "/"+y)
    img2copy=img2.copy().resize((200,400))
    img_w, img_h = img2copy.size
    offset = ((bg_w - img_w) // 2, (bg_h - img_h) // 2)
    img2copycopy=img2copy.copy()
    img2copycopy.putalpha(130)
    img1copy.paste(img2copycopy,offset,img2copy)
    img1copy.save("./content/sample_data/merged_images/"+x.split(".")[0]+y.split(".")[0]+".png", "PNG")
```
